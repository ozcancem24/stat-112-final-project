```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("mobile_price_synthetic_2015_2024_DIRTY.csv")
```

```python
df.head()
```

```python
df["Has5G"] = df["Has5G"].astype(str).str.lower().str.strip()
df["Has5G"] = df["Has5G"].str.extract(r"(yes|no)", expand=False)
df["Has5G"] = df["Has5G"].fillna("nan")
df["Has5G"].value_counts()
```

```python
df["Dual Sim"] = df["Dual Sim"].astype(str).str.lower().str.strip()
df["Dual Sim"] = df["Dual Sim"].str.extract(r"(yes|no)", expand=False)
df["Dual Sim"] = df["Dual Sim"].fillna("nan")  # Keep actual NaNs as "nan" string if needed
df["Dual Sim"].value_counts()
```

```python
df["Network Type"] = df["Network Type"].astype(str).str.lower().str.strip()
df["Network Type"] = df["Network Type"].str.extract(r"(4g|5g|3g)", expand=False)
df["Network Type"] = df["Network Type"].fillna("nan")
df["Network Type"].value_counts()
```

```python
df["OS"] = df["OS"].astype(str).str.lower().str.strip()
df["OS"] = df["OS"].str.replace(r'[^a-z]', '', regex=True) # Remove non-letter characters
df["OS"] = df["OS"].str.extract(r"(android|ios|harmonyos)", expand=False)
df["OS"] = df["OS"].fillna("nan")
df["OS"].value_counts()
```

```python
df["Model Tier"] = df["Model Tier"].astype(str).str.lower().str.strip()
df["Model Tier"] = df["Model Tier"].str.replace(r'[^a-z]', '', regex=True)
df["Model Tier"] = df["Model Tier"].str.extract(r"(midrange|budget|flagship)", expand=False)
df["Model Tier"] = df["Model Tier"].fillna("nan")
df["Model Tier"].value_counts()
```

```python
df["Brand"] = df["Brand"].astype(str).str.lower().str.strip()
df["Brand"] = df["Brand"].str.replace(r'[^a-z]', '', regex=True)
df["Brand"] = df["Brand"].str.extract(r"(samsung|xiaomi|apple|oneplus|oppo|google|motorola|huawei|nokia|sony|vivo|sony)", expand=False)
df["Brand"] = df["Brand"].fillna("nan")
df["Brand"].value_counts()
```
