```python
import numpy as np
import pandas as pd

df = pd.read_csv("mobile_price_synthetic_2015_2024_DIRTY.csv")
df.head()
```

```python
df.dropna(subset=['Year'], inplace=True) # Remove rows with missing 'Year' values because they cannot be filled meaningfully
df['Year'] = df['Year'].astype(int)
df['Year'].unique()
```

```python
df['Brand'].unique()
```

```python
df['Brand'] = df['Brand'].str.replace(r'[^a-zA-Z\s]', '', regex=True).str.strip().str.lower() # Fixing brand names
```

```python
duzeltmeler = {
    'aple': 'apple', 'appl': 'apple', 'appe': 'apple', 'pple': 'apple',
    'samsun': 'samsung', 'smsung': 'samsung', 'sasung': 'samsung', 'samung': 'samsung', 'amsung': 'samsung',
    'xiaoi': 'xiaomi', 'xiaom': 'xiaomi', 'xiami': 'xiaomi', 'xiomi': 'xiaomi',
    'googe': 'google', 'gogle': 'google', 'goole': 'google',
    'onepus': 'oneplus', 'onplus': 'oneplus', 'neplus': 'oneplus',
    'huawi': 'huawei', 'huwei': 'huawei', 'huaei': 'huawei', 'hawei': 'huawei',
    'noka': 'nokia', 'noia': 'nokia', 'nkia': 'nokia', 'okia': 'nokia',
    'motorla': 'motorola', 'motorola': 'motorola', 'otorola': 'motorola',
}

df['Brand'] = df['Brand'].replace(duzeltmeler)
```

```python
df['Brand'] = df['Brand'].str.capitalize()
df['Brand'].unique()
```

```python
kalan_duzeltmeler = {
    'Ppo': 'Oppo',
    'Viv': 'Vivo',
    'Motoola': 'Motorola'
}
df['Brand'] = df['Brand'].replace(kalan_duzeltmeler)

df.dropna(subset=['Brand'], inplace=True)

df['Brand'].unique()
```

```python
df['Model Tier'] = df['Model Tier'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True).str.strip().str.lower() # Fixing Model Tier names
```

```python
df['Model Tier'].unique()
```

```python
tier_duzeltme = {
    'flagsip': 'Flagship', 'flgship': 'Flagship', 'fagship': 'Flagship',
    'flaghip': 'Flagship', 'flagshp': 'Flagship', 'flagship': 'Flagship',
    'idrange': 'Midrange', 'midrnge': 'Midrange', 'midrang': 'Midrange',
    'mdrange': 'Midrange', 'midrane': 'Midrange', 'midrage': 'Midrange',
    'mirange': 'Midrange', 'midrange': 'Midrange',
    'budet': 'Budget', 'budge': 'Budget', 'budgt': 'Budget',
    'buget': 'Budget', 'udget': 'Budget', 'budget': 'Budget',
    'bdget': 'Budget'
}

df['Model Tier'] = df['Model Tier'].replace(tier_duzeltme)
df.dropna(subset=['Model Tier'], inplace=True)
df['Model Tier'].unique()
```

```python
df = df.copy() # Ensure we are working on a copy to avoid SettingWithCopyWarning
df["Has5G"] = df["Has5G"].astype(str).str.lower().str.strip() 
df["Has5G"] = df["Has5G"].str.extract(r"(yes|no)", expand=False)
df = df.dropna(subset=["Has5G"]) # Remove rows where Has5G could not be determined
df["Has5G"].value_counts()
```

```python
df["Dual Sim"] = df["Dual Sim"].astype(str).str.lower().str.strip()
df["Dual Sim"] = df["Dual Sim"].str.extract(r"(yes|no)", expand=False)
df = df.dropna(subset=["Dual Sim"]) # Remove rows where Dual Sim could not be determined
df["Dual Sim"].value_counts()
```

```python
df["Network Type"] = df["Network Type"].astype(str).str.lower().str.strip()
df["Network Type"] = df["Network Type"].str.extract(r"(4g|5g|3g)", expand=False)
df.loc[df["Has5G"] == "yes", "Network Type"] = "5g" # If Has5G is yes, Network Type must be 5g
df["Network Type"] = df["Network Type"].fillna("4g") # Assuming missing means 4g because it is the most common one
df["Network Type"].value_counts()
```

```python
df["OS"] = df["OS"].astype(str).str.lower().str.strip()
df["OS"] = df["OS"].str.replace(r'[^a-z]', '', regex=True) # Remove non-letter characters
df["OS"] = df["OS"].str.extract(r"(android|ios|harmonyos)", expand=False)
df = df.drop("OS", axis=1)
```

```python
df = df.dropna(subset=["Year","Brand","Model Tier"]) # Final check to ensure no critical missing values remain
```

```python
df['CPU_GHz'] = df['CPU_GHz'].fillna(df['CPU_GHz'].mean()) #Mean was used because 'CPU_GHz' is continuous numerical variable.
```

```python
df['Resolution_MP'] = df['Resolution_MP'].fillna(df['Resolution_MP'].mean()) #Mean was used because 'Resolution_MP' is continuous numerical variable.
df['Resolution_MP'] = df['Resolution_MP'].round(2) #Values were rounded because there were too many decimal places.
```

```python
df['PriceClass'] = df['PriceClass'].fillna(df['PriceClass'].median()) #Median was used because 'Price_Class' is ordinal categorical variable.
df["PriceClass"] = df["PriceClass"].astype(int) #For Price_Class, float was changed to integer because 'Price_Class is a categorical variable. It was float only because of missing values earlier.
```

```python
df['Battery_mAh'] = df['Battery_mAh'].fillna(df["Battery_mAh"].mean()) #Mean was used because 'Battery_mAh' is continuous numerical variable.
```

```python
df['Battery_mAh'] = df['Battery_mAh'].round(2) # Values were rounded because there were too many decimal places.
```

```python
df['ScreenSize_in'] = df['ScreenSize_in'].fillna(df['ScreenSize_in'].mean()) #Mean was used because 'ScreenSize_in' is continuous numerical variable.
df['ScreenSize_in'] = df['ScreenSize_in'].round(2) # Values were rounded because there were too many decimal places.
```

```python
df["RAM_GB"] = df["RAM_GB"].fillna(df["RAM_GB"].median()) # Median was used because 'RAM_GB' is discrete numerical variable.
```

```python
df["StorageGB"] = df["StorageGB"].fillna(df["StorageGB"].median()) # Median was used because 'Storage_GB' shows a skewed distribution.
```

```python
df["Camera_MP"] = df["Camera_MP"].fillna(df["Camera_MP"].median()) # Median was used because 'Camera_MP' shows a skewed distribution.
```

```python
df["Weight_g"] = df["Weight_g"].fillna(df["Weight_g"].mean()) # Mean was used because 'Weight_g' is a continuous numerical variable.
df['Weight_g'] = df['Weight_g'].round(2) # Values were rounded to improve readability.
```

```python
df.columns = ['Year', 'Brand', 'Model_Tier', 'Network_Type', 'Dual_Sim', 'Has5G',
       'Battery_mAh', 'ScreenSize_in', 'RAM_GB', 'Storage_GB', 'Camera_MP',
       'Weight_g', 'CPU_GHz', 'Resolution_MP', 'Price_Class'] # Standardizing column names
```

```python
df.duplicated().sum() # Checking for duplicates
```

```python
df = df.drop_duplicates() # Dropping duplicate rows
```
